```python
from transformers import pipeline
```

```python
classifier = pipeline("sentiment-analysis") # by default loading the distilbert model
```

### Imp.Points :
- By default, this pipeline uses distilbert-base-uncased-finetuned-sst-2-english

- Here, uncased means it can handle case sensitivity.

- It have same actions for "apple" and "Apple"

- sst-2 stands for Standford Sentiment Treebank-2 -> Dataset from Hugging Face

- HF_TOKEN is used to use the model which sst-2, it allows us to use that model

```python
summarization(text,max_length=10)
```

### There are 3 ways to load any model like text-generation,summarization or sentiment-analysis,etc.
1. Through API(like OPENAI API, Groq API)
2. In memory(directly)
3. In disk

> **Here, in the above code, we are loading the model directly into memory.**


```python
classifier = pipeline("zero-shot-classification")
```

```python
res = classifier(
    "This is a course about Python list comprehension",
    candidate_labels=["education", "politics", "business"]
)
```

```python
print(res)
```
